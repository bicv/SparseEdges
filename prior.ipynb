First, we retrieve edges from a prior edge extraction

```python
imageslist, edgeslist, RMSE = mp.process(exp='testing_vanilla', name_database='serre07_distractors')
```

```python
#fig, (a1, a2) = plt.subplots(1, 2)
#fig, a1 = mp.histedges_theta(edgeslist, fig=fig, a=a1)
#fig, a2 = mp.histedges_theta(edgeslist, fig=fig, a=a1)
fig_width_pt = 318.670  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inches
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig = plt.figure(figsize=(fig_width/1.618, fig_width/1.618))
ax = fig.add_subplot(111, axisbg='w', polar=True)
fig, ax = mp.histedges_theta(edgeslist, fig=fig, a=ax)
v_hist, v_theta_edges = mp.histedges_theta(edgeslist, display=False)
ax.plot((v_theta_edges[1:]+v_theta_edges[:-1])/2, np.ones_like(v_hist)*v_hist.mean(), 'r--')
ax.plot((v_theta_edges[1:]+v_theta_edges[:-1])/2+np.pi, np.ones_like(v_hist)*v_hist.mean(), 'r--')
if not(figpath==None): fig.savefig(figpath + 'firstorder_A.pdf')
```

Now, we may use this statistics:


```python
fig, [[B, C], [A, D]] = plt.subplots(2, 2, figsize=(fig_width, fig_width/1.618), sharex=True, subplot_kw={'axisbg':'w'})
v_hist, v_theta_edges = mp.histedges_theta(edgeslist, display=False)
v_theta_middles, v_theta_bin  = (v_theta_edges[1:]+v_theta_edges[:-1])/2, v_theta_edges[1]-v_theta_edges[0]
A.bar(v_theta_edges, np.hstack((v_hist, v_hist[-1])), alpha=.6, width=v_theta_bin)
plt.setp(A, yticklabels=[], xlim=[v_theta_edges[1], v_theta_edges[-1]], xticks=[0, np.pi/2], xticklabels=[r'$0$', r'$\pi/2$'], xlabel=r'$\theta$ (rad)', ylabel=r'$p(\theta)$')
#A.text(-np.pi/2*1.65, v_hist.max()*1.1, '(A)')
B.step(v_theta_edges, np.hstack((0, np.cumsum(v_hist))))
plt.setp(B, xticks=[], xlim=[v_theta_edges[0], v_theta_edges[-1]], ylim=[0,1], yticks=[0., 0.5, 1.], ylabel=r'$P(\theta)$')
#B.text(-np.pi/2*1.65, 1.05, '(B)')
z = np.linspace(.5/pe.n_theta, 1.-.5/pe.n_theta, pe.n_theta)
theta = np.interp(z, np.hstack((0, np.cumsum(v_hist))), v_theta_edges)
C.plot(v_theta_edges, np.hstack((0, np.cumsum(v_hist))))
C.hlines(z, [-np.pi/2], theta, lw=.5)
C.vlines(theta, [0], z, lw=.5, color='r')
#C.text(-np.pi/2*1.1, 1.05, '(C)')
plt.setp(C, xlim=[v_theta_edges[0], v_theta_edges[-1]], ylim=[0,1], xticks=[], yticks=[])
D.axis('tight')
D.vlines(theta, [0], np.ones_like(z), lw=.5, color='r')
D.plot(theta, np.ones_like(z), ms='^', lw=0, color='r')
plt.setp(D, ylim=[0, 1.3], yticks=[], xlim=[v_theta_edges[1], v_theta_edges[-1]], xticks=[0, np.pi/2], xticklabels=[r'$0$', r'$\pi/2$'], xlabel=r'$\theta$ (rad)')
#D.text(-np.pi/2*1.1, 1.35, '(D)')
plt.tight_layout()
if not(figpath==''): fig.savefig(figpath + 'laughlin.pdf')
```

```python
v_hist = np.load('mat/testing_vanilla_serre07_distractors_hist.npy')
print v_hist.max()
```

```python
#mp.pe.scale_invariant = False
#mp.pe.weight_by_distance = False

def P_ij(edges, v_hist):
    P = np.zeros((edges.shape[1], edges.shape[1]))
    for i in range(edges.shape[1]):
        for j in range(i+1, edges.shape[1]):
            try:
                v_hist_ = mp.cohistedges(edges[:, [i, j]][:, :, np.newaxis], display=False)
                p = (v_hist_ * v_hist).sum()
                if np.isnan(p): p = 0.
            except:
                p = 0.
            P[i, j] = p
            P[j, i] = P[i, j]
    #for i in range(edges.shape[1]):
    #    P[:, i] /= P[:, i].sum()
    #for i in range(edges.shape[1]):
    #    P[i, :] /= P[i, :].sum()
    return P
P = P_ij(edges, v_hist)
```

```python
plt.imshow(P)
print P.max(), P.sum(axis=1)
```

```python
p = np.ones((edges.shape[1],)) * edges[4, :]
for step in range(15):
    p /= p.sum()
    p = np.dot( p.T, P)
edges[4, :] = p
fig, a = mp.show_edges(edges, image=image, v_min=v_min, v_max=v_max, color='toto', show_phase=False) #
```
